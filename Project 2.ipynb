## B.1 Quantum Anharmonic Oscillator Operators

$\text{Proof for }\langle m|\hat x^2|n\rangle$

$$\hat a_+=\frac{1}{\sqrt{2}}(\hat x-i\hat p) \qquad \hat a_-=\frac{1}{\sqrt{2}}(\hat x+i\hat p)$$
$$\hat a_+|n\rangle=\sqrt{n+1}|n+1\rangle \qquad \hat a_-|n\rangle=\sqrt{n}|n-1\rangle$$
$$\hat a_+ + \hat a_- = \frac{1}{\sqrt{2}}(\hat x-i\hat p+\hat x+i\hat p)=\frac{2}{\sqrt{2}}\hat x$$
$$\hat x = \frac{1}{\sqrt{2}}(\hat a_+ + \hat a_-)$$
$$\hat x^2 = \frac{1}{2}(\hat a_+ + \hat a_-)(\hat a_+ + \hat a_-)$$
$$\hat x^2 = \frac{1}{2}(\hat a_+ \hat a_+ + \hat a_+ \hat a_- + \hat a_- \hat a_+ + \hat a_- \hat a_-)$$
$$\hat x^2 |n\rangle = \frac{1}{2}(\hat a_+ \hat a_+|n\rangle + \hat a_+ \hat a_-|n\rangle + \hat a_- \hat a_+|n\rangle + \hat a_- \hat a_-|n\rangle)$$
$$\hat x^2 |n\rangle = \frac{1}{2}(\sqrt{(n+1)(n+2)}|n+2\rangle +(2n+1)|n\rangle + \sqrt{n(n-1)}|n-2\rangle)$$
$$\langle m|\hat x^2 |n\rangle = \frac{1}{2}(\sqrt{(n+1)(n+2)}\langle m|n+2\rangle +(2n+1)\langle m|n\rangle + \sqrt{n(n-1)}\langle m|n-2\rangle)$$
$$\langle m|\hat x^2|n\rangle=(n+\frac{1}{2})\delta_{nm} + \frac{1}{2}\sqrt{(n+1)(n+2)}\,\delta_{n,m-2} + \frac{1}{2}\sqrt{n(n-1)\,}\,\delta_{n,m+2}$$

$\text{Proof for } \langle m|\hat x^4|n\rangle$
$$\hat x^2  = \frac{1}{2}(\hat a_+ \hat a_+ + \hat a_+ \hat a_- + \hat a_- \hat a_+ + \hat a_- \hat a_-)$$
$$\hat x^4  = \frac{1}{4}(\hat a_+ \hat a_+ + \hat a_+ \hat a_- + \hat a_- \hat a_+ + \hat a_- \hat a_-)(\hat a_+ \hat a_+ + \hat a_+ \hat a_- + \hat a_- \hat a_+ + \hat a_- \hat a_-)$$
$$\hat x^4  = \frac{1}{4}(\hat a_+ \hat a_+\hat a_+ \hat a_+ + \hat a_+ \hat a_+\hat a_+ \hat a_- + \hat a_+ \hat a_+\hat a_- \hat a_+ + \hat a_+ \hat a_+\hat a_- \hat a_- + \hat a_+ \hat a_-\hat a_+ \hat a_+ + \hat a_+ \hat a_-\hat a_+ \hat a_- + \hat a_+ \hat a_-\hat a_- \hat a_+ + \hat a_+ \hat a_-\hat a_- \hat a_- + \hat a_- \hat a_+\hat a_+ \hat a_+ + \hat a_- \hat a_+\hat a_+ \hat a_- + \hat a_- \hat a_+\hat a_- \hat a_+ + \hat a_- \hat a_+\hat a_- \hat a_-+\hat a_- \hat a_-\hat a_+ \hat a_+ + \hat a_- \hat a_-\hat a_+ \hat a_- + \hat a_- \hat a_-\hat a_- \hat a_+ + \hat a_- \hat a_-\hat a_- \hat a_-)$$
$$\hat x^4|n\rangle  = \frac{1}{4}(\hat a_+ \hat a_+\hat a_+ \hat a_+ + \hat a_+ \hat a_+\hat a_+ \hat a_- + \hat a_+ \hat a_+\hat a_- \hat a_+ + \hat a_+ \hat a_+\hat a_- \hat a_- + \hat a_+ \hat a_-\hat a_+ \hat a_+ + \hat a_+ \hat a_-\hat a_+ \hat a_- + \hat a_+ \hat a_-\hat a_- \hat a_+ + \hat a_+ \hat a_-\hat a_- \hat a_- + \hat a_- \hat a_+\hat a_+ \hat a_+ + \hat a_- \hat a_+\hat a_+ \hat a_- + \hat a_- \hat a_+\hat a_- \hat a_+ + \hat a_- \hat a_+\hat a_- \hat a_-+\hat a_- \hat a_-\hat a_+ \hat a_+ + \hat a_- \hat a_-\hat a_+ \hat a_- + \hat a_- \hat a_-\hat a_- \hat a_+ + \hat a_- \hat a_-\hat a_- \hat a_-)|n\rangle$$
$$\hat x^4 |n\rangle=\frac{1}{4}(\sqrt{(n+1)(n+2)(n+3)(n+4)}|n+4\rangle+n\sqrt{(n+1)(n+2)}|n+2\rangle+(n+1)\sqrt{(n+1)(n+2)}|n+2\rangle+n(n-1)|n\rangle+(n+2)\sqrt{(n+1)(n+2)}|n+2\rangle+n^2|n\rangle+n(n+1)|n\rangle+(n-2)\sqrt{n(n-1)}|n-2\rangle+(n+3)\sqrt{(n+1)(n+2)}|n+2\rangle+n(n+1)|n\rangle+(n+1)^2|n\rangle+(n-1)\sqrt{n(n-1)}|n-2\rangle+(n+1)(n+2)|n\rangle+n\sqrt{n(n-1)}|n-2\rangle+(n+1)\sqrt{n(n-1)}|n-2\rangle+\sqrt{n(n-1)(n-2)(n-3)}|n-4\rangle)\\
\hat x^4 |n\rangle=\frac{1}{4}(6n^2+6n+3)|n\rangle+(n+\frac{3}{2})\sqrt{(n+1)(n+2)}|n+2\rangle+(n-\frac{1}{2})\sqrt{n(n-1)}|n-2\rangle+\frac{1}{4}\sqrt{(n+1)(n+2)(n+3)(n+4)}|n+4\rangle+\frac{1}{4}\sqrt{n(n-1)(n-2)(n-3)}|n-4\rangle\\
\langle m|\hat x^4 |n\rangle=\frac{1}{4}(6n^2+6n+3)\langle m|n\rangle+(n+\frac{3}{2})\sqrt{(n+1)(n+2)}\langle m|n+2\rangle+(n-\frac{1}{2})\sqrt{n(n-1)}\langle m|n-2\rangle+\frac{1}{4}\sqrt{(n+1)(n+2)(n+3)(n+4)}\langle m|n+4\rangle+\frac{1}{4}\sqrt{n(n-1)(n-2)(n-3)}\langle m|n-4\rangle$$
$$\langle m|\hat x^4 |n\rangle=\frac{1}{4}(6n^2+6n+3)\delta_{nm}+(n+\frac{3}{2})\sqrt{(n+1)(n+2)}\delta_{n,m-2}+(n-\frac{1}{2})\sqrt{n(n-1)}\delta_{n,m+2}+\frac{1}{4}\sqrt{(n+1)(n+2)(n+3)(n+4)}\delta_{n,m-4}+\frac{1}{4}\sqrt{n(n-1)(n-2)(n-3)}\delta_{n,m+4}$$





### Functions for B

```python
#Computes the perturbed hamiltonian for the anharmonic oscillator
def H_L(N, L):
    n = np.arange(N)
    #The matrix elements for the different kroniger deltas are found
    H_nm = np.diag((n + 0.5) + L * 0.25 * (6 * n ** 2 + 6 * n + 3))
    H_nm2 = np.diag(L * np.sqrt((n + 1) * (n + 2)) * (n + 1.5) , 2)
    H_nm4 = np.diag(L * 0.25 * np.sqrt((n + 1) * (n + 2) * (n + 3) * (n + 4)), 4)
    
    #The matrices are added; since the hermitian is symmetric, the transpose is taken and added to get the other half of the matrix
    return H_nm + H_nm2[:N, :N] + H_nm4[:N, :N] + np.transpose(H_nm2[:N, :N] + H_nm4[:N, :N])


#Computes the hermitian polynomial values centered at x up to h_n(x)
def herm(x, n):
    #Initializes the array
    herm = np.zeros(n)
    herm[0] = 1
    herm[1] = 2 * x
    
    #Uses the algorithm for hermite polynomials to compute the remaining polynomial values
    for i in range(2, n):
        herm[i] = 2 * x * herm[i - 1] - 2 * (n - 1) * herm[i - 2]
    
    return herm


#Calculates the wavefunctions for the harmonic oscillator using the definition
def psi(x, n):
    psi = np.zeros(n)
    for N in range(n):
        psi[N] = (2 ** n * np.math.factorial(N) * np.sqrt(np.pi)) ** (-.5) * np.exp(-x ** 2 / 2)
    
    return psi * herm(x, n)
```

## B.3 $E(\lambda)$ Graphs

```python
N = 20
E = np.zeros([21, 4])

#The first four energy values are found using various lambda values to be used later when graphing
for l in range(21):
    H = H_L(N, l * 0.05)
    E[l] = r.real_eigen(H)[0][:4]
E = np.transpose(E)

#Plots are created for E vs lambda and delta E vs lambda
fig,ax1 = plt.subplots()
fig,ax2 = plt.subplots()
for i in range(4):
    ax1.plot(np.linspace(0, 1, 21),E[i], label = 'E_' + str(i + 1))
for i in range(4 - 1):
    ax2.plot(np.linspace(0, 1, 21),(E[i + 1] - E[i]), label = 'E_' + str(i + 1) + ' - ' + 'E_' + str(i))

ax1.set_title('Energy vs Lambda')
ax2.set_title('Delta Energy vs Lambda')

ax1.legend()
ax2.legend()
plt.show()
```

## B.5 Harmonic vs Anharmonic Oscillator Wavefunctions

```python
n = 4
L = 1
H = H_L(n, L)
x = np.linspace(-5, 5, 1000)
harm_psi = np.zeros((1000, n))
anharm_psi = np.zeros((1000, n))
psi0 = r.real_eigen(H)[1]

#The values for the oscillator wavefunctions are found
for i in range(1000):
    #The harmonic oscillator wavefunction is found with the defined equation from notes 6
    harm_psi[i] = psi(x[i], n)
    #The anharmonic oscillator wavefunctions are found by dotting the eigenvectors with their respective
    #harmonic wavefunction and the values are saved in a matrix
    for j in range(4):
        anharm_psi[i][j] = np.dot(psi(x[i], n), psi0[:, j])

#Plots are created to compare the harmonic oscillator wavefunctions with the anharmonic oscillator wavefunctions
fig, ax5 = plt.subplots()
fig, ax6 = plt.subplots(2, 2, figsize = (10, 10))

ax5.plot(x, harm_psi[:, :4], label='Psi_' + str(i))
ax5.set_title('Original Wavefunctions')

ax6[0][0].plot(x, anharm_psi[:, 0], label = 'Anharmonic')
ax6[0][0].plot(x, harm_psi[:, 0], label = 'Harmonic')
ax6[0][0].set_title('Psi_0')

ax6[0][1].plot(x, anharm_psi[:, 1], label = 'Anharmonic')
ax6[0][1].plot(x, harm_psi[:, 1], label = 'Harmonic')
ax6[0][1].set_title('Psi_1')

ax6[1][0].plot(x, anharm_psi[:, 2], label = 'Anharmonic')
ax6[1][0].plot(x, harm_psi[:, 2], label = 'Harmonic')
ax6[1][0].set_title('Psi_2')

ax6[1][1].plot(x, anharm_psi[:, 3], label = 'Anharmonic')
ax6[1][1].plot(x, harm_psi[:, 3], label = 'Harmonic')
ax6[1][1].set_title('Psi_3')

ax6[0][0].legend()
ax6[0][1].legend()
ax6[1][0].legend()
ax6[1][1].legend()
plt.show()
```
